```python
import numpy as np
from scipy import signal
import matplotlib.pyplot as plt
%matplotlib inline

duration = 20.0    # seconds
amplitude = 0.5
frequency = 440

sample_rate = 200 # sample pr. second
n_samples = int(duration * sample_rate)
time_low = np.linspace(0, duration, n_samples) # time variable
sound_low = amplitude * np.sin(2 * np.pi * frequency * time_low) # pure sine wave at 440 Hz
# sound_low += np.random.random(len(sound_low))

yf = np.fft.fft(sound_low)
xf = np.linspace(0.0, sample_rate / 2, n_samples / 2)
plt.plot(xf, abs(yf[:len(xf)]))

plt.figure()
sf = 2 / duration * yf**2
plt.plot(xf, abs(sf[:len(xf)]))

plt.figure()
xf, yf = signal.periodogram(sound_low, sample_rate)
plt.plot(xf, yf)
```
